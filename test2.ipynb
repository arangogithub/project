```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
#rint = 0.2
#const = (rint**2-1)/(8*np.log(rint))
a, b, c = 6, 5, 1
def u(x,y):
    return (1-(x**2+y**2))/4 + c*np.log(((x-a)**2+y**2))+ c*np.log(((y-b)**2+x**2)) 

#def Dux(x,y): 
#return -x/2+ 2*x/(x**2+y**2)
```

```python
wd = 8
N = 100
xx, yy  = np.linspace(-wd,wd,N), np.linspace(-wd,wd,N)
X, Y = np.meshgrid(xx, yy)
U = u(X,Y)
```

```python
plt.figure()
levels = [-0.5, 0.0000001, 0.5]
colores = ('Blue', 'Black','Red' )
line_widths = (1,2,1)
plt.contour(X,Y, U, levels, colors = colores,                     # levels lot plot
                 linewidths = line_widths           # line widths             
                 )
plt.plot(a,0,'bo', markersize=7)
plt.plot(0,b,'bo', markersize=7)
plt.show()
```
