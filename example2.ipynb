```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import init_printing
from sympy import *
```

```python
x, y, t = symbols('x y t')
eps = Rational(2,5)
inrad = Rational(1,10) 
cnst = simplify((inrad**2-1)/(8*log(inrad)))
exprs_u = (1-(x**2+y**2))/4 + cnst*sp.log((x-eps)**2+y**2)
u = lambdify((x, y), exprs_u, "numpy") 
```

```python
detm = simplify(diff(exprs_u,x,x)*diff(exprs_u,y,y) - diff(exprs_u,x,y)**2)
jacD = lambdify((x, y), detm, "numpy") 
```

```python
exprs_du_theta = diff(exprs_u,x)*sp.cos(t) + diff(exprs_u,y)*sp.sin(t)
du_t = lambdify((t,x, y), exprs_du_theta, "numpy")  
```

```python
simplify(exprs_du_theta)
```

```python
simplify(diff(exprs_u, x,x)+ diff(exprs_u, y,y))
  
```

```python
wd = 0.25
xx, yy  = np.linspace(-1-wd,1+wd,571), np.linspace(-1-wd,1+wd,571)
X, Y = np.meshgrid(xx, yy)
U = u(X,Y)
Z = du_t(0.0, X,Y)
ZA = du_t(np.pi/2, X,Y)
D = jacD(X,Y)
```

```python

plt.figure()
levels = [0.0]

plt.contour(X,Y, U, levels,                     # levels lot plot
                 linewidths=3,            # line widths
                 colors = 'Black'                  # line widths
                 )

plt.contour(X,Y, Z, levels,                     # levels lot plot
                 linewidths=2,                    # line widths
                 colors = 'Indigo',
                 alpha = 0.5
                 )

plt.contour(X,Y, ZA, levels,                     # levels lot plot
                 linewidths=3.5,                    # line widths
                 colors = 'ForestGreen',
                 alpha = 0.5
                 )

plt.contour(X,Y, D, levels,                     # levels lot plot
                 linewidths=2,            # line widths
                 colors = 'Red',  
                 alpha = 0.5
                 )


plt.title('Contour Plot u ')        # title

plt.show()
```

```python
manual_locations = [(0.5, 0.0), (-0.5, -0.0), (0, 0.5), (0.5, 0.5)]

line_colours = ('BlueViolet', 'Crimson', 'ForestGreen', 'Indigo')
line_widths = (1, 1.5, 2, 2.5)
CS = plt.contour(X,Y, Z, 4,                     # levels lot plot
                 linewidths=line_widths,            # line widths
                 colors = line_colours                  # line widths
                 )
plt.clabel(CS, inline=1,                            # add labels
          fontsize=10,                             # label font size 
          manual=manual_locations)                 # label locations
```
