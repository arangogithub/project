```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
rint = 0.2
const = (rint**2-1)/(8*np.log(rint))

def u(x,y):
    return (1-(x**2+y**2))/4 + const*np.log((x**2+y**2))

def Dux(x,y): 
    return -x/2+ 2*x/(x**2+y**2)
```

```python
wd = 0.0
N = 100
xx, yy  = np.linspace(-1-wd,1+wd,N), np.linspace(-1-wd,1+wd,N)
X, Y = np.meshgrid(xx, yy)
U = u(X,Y)
D = Dux(X,Y)
```

```python
D
```

```python
D = np.ma.array(D)
D[U<0.00] = np.ma.masked
D
```

```python
plt.figure()
levels = [-0.5, 0.0000001, 0.5]
line_widths = (3,2,1)
plt.contour(X,Y, D, levels,                     # levels lot plot
                 linewidths = line_widths,            # line widths
                 colors = 'Black'                  # line widths
                 )

levels = [0.0000001, 0.05]
plt.contour(X,Y, U, levels,                     # levels lot plot
                 linewidths = line_widths,            # line widths
                 colors = 'Black'                  # line widths
                 )

plt.show()
```
