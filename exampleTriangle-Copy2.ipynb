```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import init_printing
from sympy import *
```

```python
x, y, t = symbols('x y t')
a, l  = np.sqrt(3), 1
exprs_u = Rational(1,12)*(y+1)*((y-2)**2-3*x**2)+ Rational(1,2)*x
u = lambdify((x, y), exprs_u, "numpy") 
```

```python
detm = simplify(diff(exprs_u,x,x)*diff(exprs_u,y,y) - diff(exprs_u,x,y)**2)
laplacian = simplify(diff(exprs_u,x,x)+diff(exprs_u,y,y))
jacD = lambdify((x, y), detm, "numpy") 
```

```python
exprs_du_theta = diff(exprs_u,x)*sp.cos(t) + diff(exprs_u,y)*sp.sin(t)
du_t = lambdify((t,x, y), exprs_du_theta, "numpy")  
```

```python
curv =  -(diff(exprs_u,x,x)*diff(exprs_u,y)**2 + diff(exprs_u,y,y)*diff(exprs_u,x)**2  -
         2*diff(exprs_u,x,y)*diff(exprs_u,x)*diff(exprs_u,y))/(diff(exprs_u,x)**2+diff(exprs_u,y)**2)**(Rational(3,2))  
kappa =   lambdify((x, y), curv, "numpy")    
#/(diff(exprs_u,x)**2+diff(exprs_u,y)**2)**(Rational(3,2))
```

```python
init_printing(use_latex='mathjax')
laplacian
```

```python
wd = 2
N = 500
xx, yy  = np.linspace(-wd,wd,N), np.linspace(-wd,wd,N)
X, Y = np.meshgrid(xx, yy)
U = u(X,Y)
DU1 = du_t(-np.pi/2, X,Y)
DU11 = du_t(-np.pi/4, X,Y)
DU2 = du_t(np.pi/3, X,Y)
DU22 = du_t(np.pi/6, X,Y)
DU33 = du_t(np.pi, X,Y)
DU44 = du_t(np.pi/2.3, X,Y)
DU55 = du_t(np.pi/2.45, X,Y)
Jdet = jacD(X,Y)
K = kappa(X,Y)
```

```python

plt.figure()


levels = [0.000]
plt.contour(X,Y, DU1, levels,                     # levels lot plot
                 linewidths=3,                    # line widths
                 colors = 'Indigo',
                 alpha = 0.5
                 )

levels = [0.000]
plt.contour(X,Y, DU11, levels,                     # levels lot plot
                 linewidths=1,                    # line widths
                 colors = 'Indigo',
                 alpha = 0.5
                 )

plt.contour(X,Y, DU2, levels,                     # levels lot plot
                 linewidths=3.5,                    # line widths
                 colors = 'ForestGreen',
                 alpha = 0.5
                 )
plt.contour(X,Y, DU22, levels,                     # levels lot plot
                 linewidths=3.5,                    # line widths
                 colors ='BlueViolet',
                 alpha = 0.5
                 )

plt.contour(X,Y, DU33, levels,                     # levels lot plot
                 linewidths=1.5,                    # line widths
                 colors ='BlueViolet',
                 alpha = 0.5
                 )

plt.contour(X,Y, DU44, levels,                     # levels lot plot
                 linewidths=1.5,                    # line widths
                 colors ='ForestGreen',
                 alpha = 0.5
                 )

plt.contour(X,Y, DU55, levels,                     # levels lot plot
                 linewidths=1.5,                    # line widths
                 colors ='Yellow',
                 alpha = 0.5
                 )

plt.contour(X,Y, Jdet, levels,                     # levels lot plot
                 linewidths=1,            # line widths
                 colors = 'Red',  
                 alpha = 0.75
                 )

plt.title('Contour Plot u ')        # title

plt.show()
```
