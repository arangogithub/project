```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import init_printing
from sympy import *
```

```python
x, y, t = symbols('x y t')
eps = 6+ Rational(1,2)
inrad = Rational(2,10) 
cnst = simplify((inrad**2-1)/(8*sp.log(inrad)))
exprs_u = (1-(x**2+y**2))/4 + cnst*sp.log((x-eps)**2+y**2)
u = lambdify((x, y), exprs_u, "numpy") 
```

```python
detm = simplify(diff(exprs_u,x,x)*diff(exprs_u,y,y) - diff(exprs_u,x,y)**2)
jacD = lambdify((x, y), detm, "numpy") 
```

```python
exprs_du_theta = diff(exprs_u,x)*sp.cos(t) + diff(exprs_u,y)*sp.sin(t)
du_t = lambdify((t,x, y), exprs_du_theta, "numpy")  
```

```python
curv =  -(diff(exprs_u,x,x)*diff(exprs_u,y)**2 + diff(exprs_u,y,y)*diff(exprs_u,x)**2  -
         2*diff(exprs_u,x,y)*diff(exprs_u,x)*diff(exprs_u,y))/(diff(exprs_u,x)**2+diff(exprs_u,y)**2)**(Rational(3,2))  
kappa =   lambdify((x, y), curv, "numpy")    
#/(diff(exprs_u,x)**2+diff(exprs_u,y)**2)**(Rational(3,2))
```

```python
init_printing(use_latex='mathjax')
curv
```

```python
wd = 0.25
xx, yy  = np.linspace(-1-wd,1+wd,900), np.linspace(-1-wd,1+wd,900)
X, Y = np.meshgrid(xx, yy)
U = u(X,Y)
DU1 = du_t(0.0, X,Y)
DU2 = du_t(np.pi/2, X,Y)
Jdet = jacD(X,Y)
K = kappa(X,Y)
```

```python

plt.figure()
levels = [0.0, 0.087]
line_widths = (3,1)
plt.contour(X,Y, U, levels,                     # levels lot plot
                 linewidths = line_widths,            # line widths
                 colors = 'Black'                  # line widths
                 )

levels = [0.000]
plt.contour(X,Y, DU1, levels,                     # levels lot plot
                 linewidths=2,                    # line widths
                 colors = 'Indigo',
                 alpha = 0.5
                 )

plt.contour(X,Y, DU2, levels,                     # levels lot plot
                 linewidths=3.5,                    # line widths
                 colors = 'ForestGreen',
                 alpha = 0.5
                 )

plt.contour(X,Y, Jdet, levels,                     # levels lot plot
                 linewidths=1,            # line widths
                 colors = 'Red',  
                 alpha = 0.75
                 )

plt.contour(X,Y, K, levels,                     # levels lot plot
                 linewidths=4,            # line widths
                 colors = 'Crimson',  
                 alpha = 0.75
                 )


plt.title('Contour Plot u ')        # title

plt.show()
```
